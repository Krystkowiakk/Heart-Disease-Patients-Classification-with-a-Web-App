# Heart Disease Indicators app
## Metis-Project-7-Data Engineering

Krystian Krystkowiak 2022

**Abstract:**

The goal of this project was to build Web app that can help  identifying high-risk Heart Disease patients. I used data from Kaggle, stored in **SQL**, processed and deployed at **Streamlit**. I visualised and communicated my findings during 5 minutes presentation.

**Design:**

Our potential clients can be medical institutions. Also wide range of businesses where heart disease may be concern: insurances, medical apps, fitness or nutritionist. But app itself can be particularry use by **individuals conscious about health**.

The dataset that originally comes from the CDC and is a major part of the **Behavioural Risk Factor Surveillance System** (BRFSS), which conducts annual telephone surveys to gather data on the health status of U.S. residents. BRFSS completes more than 300,000 adult interviews each year, making it the largest continuously conducted health survey system in the world.". The most recent dataset (as of February 15, 2022) includes data from 2020. 

Our app Web app that can **help increase awareness** and raise red flag during initial questionnaire.
