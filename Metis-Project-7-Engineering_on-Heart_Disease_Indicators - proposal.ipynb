# Heart Disease Indicators app 
## Metis Project 7 - Engineering on Heart Disease Indicators - project proposal
Krystian Krystkowiak 2022

**Question/Need:**

Our clients are doctors and medical institutions. Also wide range of businesses where heart disease may be concern: insurances, fitness or nutritionist. It could be also use by individuals conscious about health.

The **web app** that can help to **indentyfy factors** that have greatest impact on **heart disease**, predict a patient's condition and raise red flag during initial questionnaire.

**Data Description:**

I intend to use of the dataset that originally comes from the CDC and is a major part of the Behavioural Risk Factor Surveillance System (BRFSS), which conducts annual telephone surveys to gather **data on the health status of U.S. residents**. BRFSS completes more than 400,000 adult interviews each year, making it the largest continuously conducted health survey system in the world.". The most recent dataset (as of February 15, 2022) includes data from 2020. The dataset is initially cleaned by Kamil Pytlak, who shares it and updates yearly it at **Kaggle**:

www.kaggle.com/datasets/kamilpytlak/personal-key-indicators-of-heart-disease

Over **319,000 observations**. Individual row of data has **19 columns**. Our focus will be:

 - **HeartDisease (target)**
 - BMI
 - Smoking
 - AlcoholDrinking
 - Stroke
 - AlcoholDrinking
 - PhysicalHealth
 - MentalHealth
 - DiffWalking
 - Sex
 - AgeCategory
 - Race
 - Diabetic
 - PhysicalActivity
 - GenHealth
 - SleepTime
 - Asthma
 - KidneyDisease
 - SkinCancer
